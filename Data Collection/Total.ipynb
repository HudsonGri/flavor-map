```python
import pandas as pd
import os

total = 0
res = 0
ex = []
for filename in os.listdir('cities'):
    if 'csv' in filename:
        df = pd.read_csv(f'cities/{filename}')
        count = df['user_ratings_total'].sum()
        rows = len(df.axes[0])
        cols = len(df.axes[1])
        rowcol = rows*cols
        total+=rowcol
        temp = {}
        temp['city'] = filename.split('.')[0]
        temp['mean_rating_count'] = round(df['user_ratings_total'].mean(),5)
        temp['mean_rating'] = round(df['rating'].mean(), 5)
        temp['mean_price_level'] = round(df['price_level'].mean(),5)
        ex.append(temp)
total
```
